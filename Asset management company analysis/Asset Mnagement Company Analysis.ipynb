### 9. Create a pivot table to compare the mean and median values for ‘raised_amount_usd’ and ‘funding_round_type’ across different funding round categories..

```python
#setting the floating point to display 4 decimals
pd.options.display.float_format = '{:,.4f}'.format 

# replacing the missing values to zero
round2["raised_amount_usd"].fillna(0,inplace=True) 

#converting datatype from float to int
round2["raised_amount_usd"] = round2["raised_amount_usd"].astype(int) 

# creating pivot table to compare mean and median values
pd.pivot_table(round2, index = ['funding_round_type'], aggfunc=['mean','median']) 
```

**Explanation:** 
- `pd.options.display.float_format = '{:,.4f}'.format` - sets the floating point format to display 4 decimal places for any subsequent output.
- `round2["raised_amount_usd"].fillna(0,inplace=True)` -  replaces any missing values in the raised_amount_usd column with 0.
- `round2["raised_amount_usd"] = round2["raised_amount_usd"].astype(int)` - converts the raised_amount_usd column to integer type.
- `pd.pivot_table(round2, index = ['funding_round_type'], aggfunc=['mean','median'])` - creates a pivot table that groups the round2 DataFrame by funding_round_type and calculates the mean and median of raised_amount_usd for each group.

```python
pd.pivot_table(round2, index = ['funding_round_type'], aggfunc=['mean','median']).plot(kind="bar",figsize=(8,6))
plt.title('Mean and Median of Funding Amount by Funding Round Type') # setting the title
plt.xlabel('Funding Round Type')                                     # setting the label for x axis
plt.ylabel('Funding Amount (USD)')                                   # setting the label for y axis 

handles, labels = plt.gca().get_legend_handles_labels()             # retrieving the handles and legends of the current axis
new_labels = ['Mean Funding Amount', 'Median Funding Amount']       # declaring new handles 
plt.legend(handles, new_labels)                                     # setting the updated legend
plt.show()                                                          # displaying the plot
```

**Explanation:**

 - `handles, labels = plt.gca().get_legend_handles_labels()` -  retrieves the handles and labels of the legend of the current axis (gca stands for "get current axis").
 - `new_labels = ['Mean Funding Amount', 'Median Funding Amount']` - creates a new list of labels to replace the original labels.
 - `plt.legend(handles, new_labels)`- updates the legend of the plot with the new labels. The handles argument is a list of artists that represent the data being plotted, and the new_labels argument is a list of labels to use for each artist.
 - `plt.show()`- displays the plot with the updated legend.

- We use the `'pd.pivot_table'` function to analyze the numerical data against any particular categorical columns.Here, we have tried to analyze the mean and median values 'funding_round_type' column can take. Also, since there is just one numeric column in our dataset we need not specify the column name in 'values' parameter.  <br>

### 13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical column as followa) If the amount is less than 5 Millions then recode as TypeA. b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB. c) If the amount is greater than 7 Millions then recode as TypeC

```python
master_frame[["raised_amount_usd"]].head()
def raised_amt(r):
    if r<=5000000:
        return "Type A"
    elif r<7000000:
        return "Type B"
    else:
        return "Type C"
master_frame["raised_amount_type"]=master_frame["raised_amount_usd"].apply(raised_amt)   
master_frame.head()
```

**Explanation** - A user defined function `'raised_amt'` was made to divide the raised_amount_usd into `three categories` based on the amount. The function is then called and applied in a new column,'raised_amount_type'.

- `'raised_amt'` - takes a single argument "r" and returns a string based on the value of "r". If "r" is less than or equal to 5,000,000 the function returns "type A". If "r" is less than 7,000,000 the function returns "type B". Otherwise, the function returns "type C".

- Then we apply the "raised_amt" function to the "raised_amount_usd" column using the apply() function and created a new column called "raised_amount_type" to store the results.

**Inference**

The above analysis helps us to find out the `top investors`. We have categorised the companies investing the maximum amount under `Type C`, implying thet invest the maximum amount.

```python
count_by_raised_amount_type = master_frame.groupby('raised_amount_type')['name'].count()

# Create a bar plot of the data
count_by_raised_amount_type.plot(kind='bar', figsize=(8,6))
plt.xlabel('Raised Amount Type')
plt.ylabel('Number of Companies')
plt.title('Count of Companies by Raised Amount Type')

# Show the plot
plt.show()
```

**Explanation:** 
- `pd.options.display.float_format = '{:,.4f}'.format` - sets the floating point format to display 4 decimal places for any subsequent output.
- `round2["raised_amount_usd"].fillna(0,inplace=True)` -  replaces any missing values in the raised_amount_usd column with 0.
- `round2["raised_amount_usd"] = round2["raised_amount_usd"].astype(int)` - converts the raised_amount_usd column to integer type.
- `pd.pivot_table(round2, index = ['funding_round_type'], aggfunc=['mean','median'])` - creates a pivot table that groups the round2 DataFrame by funding_round_type and calculates the mean and median of raised_amount_usd for each group.

```python
# Filter out the 'venture' funding round type
df_venture = master_frame[master_frame.funding_round_type == 'venture']


# Group the investments by country and calculate the total investment
total_investment = df_venture.groupby('country_code')['raised_amount_usd'].sum()


# Sort the countries by the total investment and get the top 9
top_9_countries = total_investment.sort_values(ascending=False)[:9]
top_9_countries
```

**Explanation**:  <br>
We filter the master_frame DataFrame to only contain rows where the funding_round_type is "venture" and assigned the filtered DataFrame to a new variable df_venture. Then we grouped the df_venture DataFrame by country_code column and calculated the sum of raised_amount_usd for each group. The result is a pandas Series containing the total investment made in each country.
<br>
Then sorting and slicing the countries by the total investment made in descending order gave us the top 9 countries with the highest investment.
